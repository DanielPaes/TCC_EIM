```python
import numpy as np
import pandas as pd
from skimage import io
import matplotlib.pyplot as plt
from skimage import img_as_ubyte
from skimage.feature import greycomatrix, greycoprops
%matplotlib inline
```

```python
data = pd.read_csv('Diretório/arquivo.csv')
data
```

```python
glcm_im = pd.DataFrame()


# Leitura das imagens recortadas, passando para escala de cinza e reuzindo número de levels para 86.

for i in range(0, len(data['Imagem'])): 
    ent = '/home/daniel/Área de Trabalho/UFPR/TCC/Imagens/Pinus/' + str(i) + '.jpg'
    img = io.imread(ent, as_gray= True)
    imagem = img_as_ubyte(img)
    df_levels = pd.DataFrame(imagem)
    df_levels = df_levels/3
    df_levels = df_levels.apply(np.int64)
    img = df_levels.values
    
    # Aplicação do GLCM com 1 distãncias e 4 ângulos
    
    glcm = greycomatrix(img, [1],[0, np.pi/2, np.pi/4 , 3*np.pi/4], levels = 86) # np.pi/4 , 3*np.pi/4

    res_total = pd.DataFrame()
    
    # Utilização de 6 características extraídas pelo GLCM

    features = ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'ASM']

    
    
    for f in features:
        
        res_ind = greycoprops(glcm, f)
        res_ind = pd.DataFrame(res_ind, index = [i], columns = [f + '_0', f + '_1', f + '_2', f + '_3']) # f + '_2', f + '_3'
        res_total = pd.concat([res_total, res_ind],  axis=1)

    
    glcm_im = pd.concat([glcm_im, res_total], axis=0)
    

glcm_im = glcm_im.reset_index()

glcm_im
```

```python
# Criando DataFrame com as features e rótulos

glcm_im = glcm_im.rename(columns={'index':'Imagem'})

#rotulo = pd.read_csv('/home/daniel/Área de Trabalho/UFPR/TCC/DataSet/defeitos_numerados.csv')

df = pd.concat([glcm_im, data[['Defeito', 'Defeito_num']]], axis = 1)
#df = pd.concat([r, rotulo[['Defeito_num', 'Defeito', 'Defeito_bin' ]]], axis = 1)
```

```python
df.to_csv('/home/daniel/Área de Trabalho/UFPR/TCC/Resultados/Pinus_GLCM_4_Classes.csv')
```
